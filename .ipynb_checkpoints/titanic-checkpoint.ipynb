```python
import numpy as np
import tensorflow.compat.v1 as tf
```

```python
label="Survived"
features = ["Pclass","fares","Single", "Nuclear","Sex_encode","Embarked_C", "Embarked_S", "Embarked_Q",
                 "young&women","man"]
```

```python
def truefalse(data):
    if data==False:
        return 0
    elif data==True:
        return 1
    else:
        return data
    
```

```python
x_data=train[features]
y_data=train[label]
x_data.head()
for feat in features:
    x_data[feat]=x_data[feat].apply(truefalse)
x_data.to_csv("C:/Users/User/Downloads/titanic/x_data.csv")
x_data=np.loadtxt("C:/Users/User/Downloads/titanic/x_data.csv", delimiter=',', skiprows=1, unpack=True)
x_data=x_data[1:,:]
x_data
```

```python
x_data=x_data.to_csv(header=None,index=False)
y_data=y_data.to_csv(header=None,index=False)
print(type(x_data))
```

```python
X=tf.placeholder(tf.float32, shape=[None,10])
Y=tf.placeholder(tf.float32, shape=[None,1])
```

```python
W=tf.Variable(tf.random_normal([10,1]), name='weight')
b=tf.Variable(tf.random_normal([1]), name='bias')
```

```python
hypothesis=tf.sigmoid(tf.matmul(X,W)+b)
cost=-tf.reduce_mean(Y*tf.log(hypothesis)+(1-Y)*tf.log(1-hypothesis))
Train=tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)
```

```python
predicted=tf.cast(hypothesis>0.5, dtype=tf.float32)
accuracy=tf.reduce_mean(tf.cast(tf.equal(predicted, Y),dtype=tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    feed={X:x_data,Y:y_data}
    for step in range(891):
        sess.run(Train,feed_dict=feed)
        if step%200==0:
            print(step,sess.run(cost,feed_dict=feed))
      
    h,c,a=sess.run([hypothesis, predicted, accuracy], feed_dict=feed)

    print("\nHypothesis: ",h,"\nCoreect (Y): ",c,"\nAccuracy: ",a)
```

```python
# scikit-learn  ensemble   RandomForestClassifier  .
from sklearn.ensemble import RandomForestClassifier
import numpy as np
# scikit-learn  model_selection   cross_val_score   .
from sklearn.model_selection import cross_val_score

# n_estimators  .
#   ,          . (300 )
#    ,   n_estimators    . (ex: 3,000 )
n_estimators = 300

#    .
#  100 .
num_epoch = 100

# hyperparameter    .
coarse_hyperparameters_list = []

# num_epoch    .
for epoch in range(num_epoch):
    # 2 100  (int)    max_depth  .
    max_depth = np.random.randint(low=2, high=100)

    # 0.1 1.0  (float)     max_features  .
    max_features = np.random.uniform(low=0.1, high=1.0)

    # Random Forest .   .
    # 1) n_estimators.  .
    # 2) max_depth.  .      .
    # 3) max_features. Feature Sampling. 0.0 ~ 1.0   ,     feature   feature .
    # 4) n_jobs.  . -1     (Core, ex: , )  .
    # 5) random_state.       .    . (   Seed Number  )
    model2 = RandomForestClassifier(n_estimators=n_estimators,
                                  max_depth=max_depth,
                                  max_features=max_features,
                                  n_jobs=-1,
                                  random_state=37)

    # cross_val_score .     .
    # 1) model.    ( Random Forest) .
    # 2) X_train. train  feature .
    # 3) y_train. train  label .
    # 4) cv. Cross Validation  (split) .  20   20 .
    # 5) scoring.   .
    # ,     (mean)   score    .
    score = cross_val_score(model2, x_train, y_train, cv=20, scoring="precision").mean()
    
    # hyperparameter    .
    hyperparameters = {
        'epoch': epoch,
        'score': score,
        'n_estimators': n_estimators,
        'max_depth': max_depth,
        'max_features': max_features,
    }

    # hyperparameter    .
    coarse_hyperparameters_list.append(hyperparameters)

    # hyperparameter   .
    print(f"{epoch:2} n_estimators = {n_estimators}, max_depth = {max_depth:2}, max_features = {max_features:.6f}, Score = {score:.5f}")

# coarse_hyperparameters_list Pandas DataFrame .
coarse_hyperparameters_list = pd.DataFrame.from_dict(coarse_hyperparameters_list)

#  coarse_hyperparameters_list score   .
coarse_hyperparameters_list = coarse_hyperparameters_list.sort_values(by="score", ascending=False)

# coarse_hyperparameters_list      .
#  (row, column)  .
print(coarse_hyperparameters_list.shape)

# coarse_hyperparameters_list  10 .
coarse_hyperparameters_list.head(10)
```

```python
# scikit-learn  ensemble   RandomForestClassifier  .
from sklearn.ensemble import RandomForestClassifier

# scikit-learn  model_selection   cross_val_score   .
from sklearn.model_selection import cross_val_score

# n_estimators  .
#   ,          . (300 )
#    ,   n_estimators    . (ex: 3,000)
n_estimators = 300

#    .
#  100 .
num_epoch = 100

# hyperparameter    .
finer_hyperparameters_list = []

# num_epoch    .
for epoch in range(num_epoch):
    # 40 100  (int)    max_depth  .
    max_depth = np.random.randint(low=10, high=57)

    # 0.6 1.0  (float)     max_features  .
    max_features = np.random.uniform(low=0.1, high=0.19)

    # Random Forest .   .
    # 1) n_estimators.  .
    # 2) max_depth.  .      .
    # 3) max_features. Feature Sampling. 0.0 ~ 1.0   ,     feature   feature .
    # 4) n_jobs.  . -1     (Core, ex: , )  .
    # 5) random_state.       .    . (   Seed Number  )
    model2 = RandomForestClassifier(n_estimators=n_estimators,
                                  max_depth=max_depth,
                                  max_features=max_features,
                                  n_jobs=-1,
                                  random_state=37)

    # cross_val_score .     .
    # 1) model.    ( Random Forest) .
    # 2) X_train. train  feature .
    # 3) y_train. train  label .
    # 4) cv. Cross Validation  (split) .  20   20 .
    # 5) scoring.   .   RMSE .
    # ,     (mean)   score    .
    score = cross_val_score(model2, x_train, y_train, cv=20, scoring="precision").mean()
    
    # hyperparameter    .
    hyperparameters = {
        'epoch': epoch,
        'score': score,
        'n_estimators': n_estimators,
        'max_depth': max_depth,
        'max_features': max_features,
    }

    # hyperparameter    .
    finer_hyperparameters_list.append(hyperparameters)

    # hyperparameter   .
    print(f"{epoch:2} n_estimators = {n_estimators}, max_depth = {max_depth:2}, max_features = {max_features:.6f}, Score = {score:.5f}")

# finer_hyperparameters_list Pandas DataFrame .
finer_hyperparameters_list = pd.DataFrame.from_dict(finer_hyperparameters_list)

#  finer_hyperparameters_list score   .
finer_hyperparameters_list = finer_hyperparameters_list.sort_values(by="score", ascending=False)

# finer_hyperparameters_list      .
#  (row, column)  .
print(finer_hyperparameters_list.shape)

# finer_hyperparameters_list  10 .
finer_hyperparameters_list.head(10)
```

```python
# best_hyperparameters   .
best_hyperparameters = coarse_hyperparameters_list.iloc[0]

# best_hyperparameters max_depth  .
#  best_max_depth   .
best_max_depth = best_hyperparameters["max_depth"]

# best_hyperparameters max_features  .
#  best_max_features   .
best_max_features = best_hyperparameters["max_features"]

# best_max_depth best_max_features .
print(f"max_depth() = {best_max_depth}, max_features() = {best_max_features:.6f}")
```

```python
best_n_estimators = 3000
model2 = RandomForestClassifier(n_estimators=best_n_estimators,
                              max_depth=best_max_depth,
                              max_features=best_max_features,
                              random_state=37,
                              n_jobs=-1)
model2
```
